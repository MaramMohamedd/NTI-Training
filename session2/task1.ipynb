```python
def reduce_depth(image , bits):
    level = 2 ** bits
    return (image //(256 // level)) * (256 // level)
```

```python
depth_4bit = reduce_depth(image , 4)
depth_8bit = reduce_depth(image , 8)
depth_2bit = reduce_depth(image , 2)
depth_1bit = reduce_depth(image , 1)
```

```python
plt.figure(figsize = (14 , 6))
plt.subplot(141)
plt.imshow(depth_4bit , cmap = "gray")
plt.title('4 bit')


plt.subplot(142)
plt.imshow(depth_8bit , cmap = "gray")
plt.title('8 bit')

plt.subplot(143)
plt.imshow(depth_2bit , cmap = "gray")
plt.title('2 bit')

plt.subplot(144)
plt.imshow(depth_1bit , cmap = "gray")
plt.title('1 bit')
```

# **Morphological operations**

```python
import cv2 as cv
import matplotlib.pyplot as plt
from cv2 import morphologyEx , erode , dilate , MORPH_OPEN , MORPH_CLOSE
import numpy as np
```

```python
image = cv.imread("morphological operation image.jpg")
gray = cv.cvtColor(image , cv.COLOR_BGR2GRAY)
```

### **i want to check first that the image 100% binary**
why ? -> easy for detecting the noise and see the effect of morphological operations

```python
_, binary = cv.threshold(image, 127, 255, cv.THRESH_BINARY)
```

```python
plt.imshow(binary)
plt.axis('off')
plt.show()
```

```python
#modifting the image 
kernal =  np.ones((3 ,3) , np.uint8) 
img_erosion = cv.erode(binary ,  kernal  , iterations= 3)
img_dilated = cv.dilate(binary ,kernal)
img_opening = cv.morphologyEx(binary  , cv.MORPH_OPEN , kernal)
img_closing = cv.morphologyEx(binary , cv.MORPH_CLOSE , kernal)
```

```python
plt.figure(figsize=(18 , 10))
plt.subplot(1 , 2 , 1)
plt.imshow(img_erosion)
plt.axis('off')
plt.title('EROSION')

plt.subplot(1 , 2 , 2)
plt.imshow(img_dilated)
plt.axis('off')
plt.title('DILATION')

plt.show()
```

```python
plt.figure(figsize=(18 , 10))
plt.subplot(1 , 2 , 1)
plt.imshow(img_opening)
plt.axis('off')
plt.title('OPENING')

plt.subplot(1 , 2 , 2)
plt.imshow(img_closing)
plt.axis('off')
plt.title('CLOSING')
plt.show()
```

# **Countouring**

```python
# from my point of view i see that the cleanest version is the opening (erosion then dilation)
cleaned_image= img_erosion.copy()
gray_image = cv.cvtColor(cleaned_image, cv.COLOR_BGR2GRAY)
```

```python
contours, _ = cv.findContours(gray_image ,  cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
```

```python
print(f"Found {len(contours)} contours")
```

# ***firstly importing the neccessary libaries***

```python
import numpy as np 
import cv2 as cv
import matplotlib.pyplot as plt
```

###  ***Loading the classifier which appropriate for the intersting object i want to detect (face) in XML format***
here is the link [face and eye haarclassifier](https://media.geeksforgeeks.org/wp-content/uploads/20250415122920890394/harr_casscade_classifiers.zip)

```python
face_detector = cv.CascadeClassifier(r'D:\Documents\NTI training\session6\haarcascade_frontalface_default.xml')
```

#### **creating the function which detects the face**

```python
def adjusted_detect_face(img):
    face_img = img.copy() #avoiding call by reference 
    face_rect = face_detector.detectMultiScale(face_img, scaleFactor=1.2, minNeighbors=3)

    for (x, y, w, h) in face_rect:
        cv.rectangle(face_img, (x, y), (x + w, y + h), (0, 255, 0),3)
        
    return face_img
```

##### ***loading the image and start showing it with the rectangle***

```python
image  = cv.imread('HALAMADRID.png.webp')
image = cv.cvtColor(image , cv.COLOR_BGR2RGB)
image_rectangle = adjusted_detect_face(image)
plt.imshow(image_rectangle)
plt.axis('off')
plt.show()
```
